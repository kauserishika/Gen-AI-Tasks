## **Task 4: ChatPromptTemplate System**

```
from langchain_core.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):
    if role == "teacher":
        system_msg = "You are a helpful teacher who explains concepts clearly."
    elif role == "interviewer":
        system_msg = "You are an interviewer who asks challenging questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker who inspires learning."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat_template.format(topic=topic)

# Example
print(get_chat_prompt("teacher", "Neural Networks"))
print(get_chat_prompt("interviewer", "Computer Networks"))
print(get_chat_prompt("motivator", "Deep Learning"))
```

## **Task 5: Input Validation Layer**

```
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Defaulting to 'formal'")
        tone = "formal"

    return audience, tone
```

## **Task 6: Prompt Generator App**

```
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} tone step by step",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} tone as a story"
    }

    if style not in templates:
        raise ValueError("Invalid style! Choose from teaching, interview, storytelling")

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=templates[style]
    )

    return prompt_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print(generate_prompt("Deep Learning", "expert", "formal", "interview"))
print(generate_prompt("Machine Learning", "intermediate", "casual", "teaching"))
```

## **Task 7: Template Reusability Test**

```
reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("NLP", "beginner", "formal"),
    ("Deep Learning", "expert", "technical")
]

for topic, audience, tone in inputs:
    print(reuse_template.format(topic=topic, audience=audience, tone=tone))
```
