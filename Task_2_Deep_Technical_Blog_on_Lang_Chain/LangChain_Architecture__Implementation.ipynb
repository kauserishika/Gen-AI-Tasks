```
from huggingface_hub import login
login("YOUR_TOKEN")

from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline

# LLM
pipe = pipeline(
    "text-generation",   # supported in your version
    model="gpt2",        # compatible model
    max_new_tokens=100
)


llm = HuggingFacePipeline(pipeline=pipe)

# tool function
def calculator(query):
    try:
        return str(eval(query))
    except:
        return "Invalid calculation"

query = "15 * 6"

# Tool execution
tool_result = calculator(query)

# LLM explanation
response = llm.invoke(f"The result of {query} is {tool_result}. Explain this in one sentence.")

print("Answer:", tool_result)
print("Explanation:", response)
```

```
from transformers import pipeline
import re

pipe = pipeline(
    "text-generation",
    model="distilgpt2",
    max_new_tokens=80,
    temperature=0.3
)

text = pipe(
    "LangChain is a framework for building applications using large language models. It helps connect models with tools and workflows. Explain it clearly:",
)[0]["generated_text"]

# ✅ Extract only first 2 sentences
sentences = re.split(r'(?<=[.!?]) +', text)
clean_output = " ".join(sentences[:2])

print(clean_output)
```
